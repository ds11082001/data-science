## –ó–∞–¥–∞—á–∞ 1.

–°—É—â–µ—Å—Ç–≤—É–µ—Ç –ø—Ä–∏–º–µ—Ç–∞, —á—Ç–æ –µ—Å–ª–∏ –ø–µ—Ä–µ–¥ –≤–∞–º–∏ –¥–æ—Ä–æ–≥—É –ø–µ—Ä–µ–±–µ–≥–∞–µ—Ç —á–µ—Ä–Ω—ã–π –∫–æ—Ç, —Ç–æ —Å–∫–æ—Ä–æ —Å–ª—É—á–∏—Ç—Å—è –Ω–µ—É–¥–∞—á–∞. 
–í—ã –∂–µ —É–∂–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ö–æ—Ä–æ—à–æ –∑–Ω–∞–µ—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∏ —Ö–æ—Ç–∏—Ç–µ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –¥–∞–Ω–Ω—É—é –ø—Ä–∏–º–µ—Ç—É. 
–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –∑–∞–¥–∞—á—É –Ω–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º —è–∑—ã–∫–µ. 

–ü—É—Å—Ç—å $X_1, ..., X_n \sim Bern(p)$ &mdash; –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω—ã–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –≥–¥–µ $X_i = 1$, –µ—Å–ª–∏ –≤ $i$-–º –∏—Å–ø—ã—Ç–∞–Ω–∏–∏ —Å–ª—É—á–∏–ª–∞—Å—å –Ω–µ—É–¥–∞—á–∞ –ø–æ—Å–ª–µ —Ç–æ–≥–æ, –∫–∞–∫ —á–µ—Ä–Ω—ã–π –∫–æ—Ç –ø–µ—Ä–µ–±–µ–∂–∞–ª –¥–æ—Ä–æ–≥—É, –∞ $p$ &mdash; –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–∞–∫–æ–≥–æ —Å–æ–±—ã—Ç–∏—è. 
–í—ã —Ö–æ—Ç–∏—Ç–µ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É $\mathsf{H}_0\colon p=1/2$ –ø—Ä–æ—Ç–∏–≤ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã $\mathsf{H}_1\colon p>1/2$. –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ–∑–Ω–∞—á–∞–µ—Ç –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ —Å–≤—è–∑–∏ –º–µ–∂–¥—É —á–µ—Ä–Ω—ã–º –∫–æ—Ç–æ–º –∏ –Ω–µ—É–¥–∞—á–µ–π, –ø–µ—Ä–≤–∞—è &mdash; —á—Ç–æ –Ω–µ—É–¥–∞—á–∞ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —á–∞—â–µ –µ—Å–ª–∏ —á–µ—Ä–Ω—ã–π –∫–æ—Ç –ø–µ—Ä–µ–±–µ–≥–∞–µ—Ç –¥–æ—Ä–æ–≥—É.

–ù–∞ —Å–µ–º–∏–Ω–∞—Ä–µ –ø–æ–ª—É—á–µ–Ω–æ, —á—Ç–æ $S = \left\{T(x) \geqslant c_\alpha\right\}$, –≥–¥–µ $T(X) = \sum\limits_{i=1}^n X_i$, —è–≤–ª—è–µ—Ç—Å—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –Ω–∞–∏–±–æ–ª–µ–µ –º–æ—â–Ω—ã–º –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —ç—Ç–∏—Ö –≥–∏–ø–æ—Ç–µ–∑. 
–ß–µ–º—É –ø—Ä–∏ —ç—Ç–æ–º —Ä–∞–≤–Ω–æ $c_\alpha$?

 –ù–∞ —Å–µ–º–∏–Ω–∞—Ä–∞—Ö –º—ã –≤—ã—è—Å–Ω–∏–ª–∏, —á—Ç–æ $$c_{\alpha} = \min_{k \in \mathbb{N}}\sum\limits_{i = 1}^{k}p_{n, 1/2}(k) \geq \alpha = $$ $ = $ `sps.binom(n, 1/2).ppf`($1 - \alpha$).

–ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –∫—Ä–∏—Ç–µ—Ä–∏–π —Ä–∞–±–æ—Ç–∞–µ—Ç.
–í–æ–∑—å–º–∏—Ç–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏–π $n$ –∏ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ $T(X)$.
–í –∫–∞–∂–¥–æ–º —Å–ª—É—á–∞–µ –Ω–∞–π–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ $c_\alpha$, –∞ —Ç–∞–∫–∂–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–æ—Ü–µ–¥—É—Ä—ã –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑ 0/1 &mdash; –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –ª–∏ $\mathsf{H}_0$ (–≤—ã–≤–æ–¥–∏—Ç–µ 1, –µ—Å–ª–∏ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è).
–û—Ñ–æ—Ä–º–∏—Ç–µ —ç—Ç–æ –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã, –º–æ–∂–Ω–æ —á–µ—Ä–µ–∑ `pandas.DataFrame`.

*–ó–∞–º–µ—á–∞–Ω–∏–µ:* –ü–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏–∑ `scipy.stats`.

```python
alpha = 0.05
p = 0.5
count_of_samples = 20
n_arr = np.array([5, 20, 100, 1000, 10000])


c_a = np.array([])
frequency_reject = np.array([])
reject_arr = np.array([np.zeros(count_of_samples)])
for n in n_arr:
    c_a = np.append(c_a, sps.binom(n, p).ppf(1 - alpha))
    t = sps.binom(n, p).rvs(size=count_of_samples)
    reject = [t > c_a[-1]]
    reject += np.zeros(len(reject))
    reject_arr = np.append(reject_arr, reject, axis=0)
    frequency_reject = np.append(frequency_reject, np.sum(reject))

frequency_reject /= count_of_samples    
reject_arr = reject_arr[1:]
reject_arr
```

```python
reject_arr = np.hstack((c_a.reshape([reject_arr.shape[0], 1]), reject_arr))
reject_arr = np.hstack((reject_arr, frequency_reject.reshape([reject_arr.shape[0], 1])))
reject_arr
```

```python
d = {f'n = {n_arr[i]}': reject_arr[i] for i in range(len(n_arr))}
data = pd.DataFrame(d)
index = np.array([f'—Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ {i + 1}' for i in range(count_of_samples)])
index = np.hstack((['$c_{\\alpha}$'], index))
index = np.hstack((index, ['—á–∞—Å—Ç–æ—Ç–∞ –æ—Ç–≤–µ—Ä–∂–µ–Ω–∏—è']))
index = pd.Index(index)
# index
data = data.set_index(index)
data
```

–¢–µ–ø–µ—Ä—å –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö $n$ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –º–æ—â–Ω–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –ø—Ä–∏ $1/2 < p < 1$ –∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05.

–§–æ—Ä–º—É–ª–∞ –¥–ª—è –º–æ—â–Ω–æ—Å—Ç–∏: $\beta_S(\theta) = P_{\theta}(T(x) \geq c_\alpha)$

```python
power = lambda p, n, c_a: sps.binom(n, p).sf(c_a)

p_grid = np.linspace(0.501, 0.999, 499)
fig = go.Figure()
for i, n in enumerate(n_arr):
    power_grid = power(p_grid, n, c_a[i])
    fig.add_trace(go.Scatter(x=p_grid, y=power_grid, mode='lines', name=f'ùëõ = {n}'))
fig.update_layout(title='–§—É–Ω–∫—Ü–∏—è –º–æ—â–Ω–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö ùëõ', 
                  xaxis_title="$p$", yaxis_title="$\\beta_{S}(p)$")
fig.show()
pof.plot(fig, filename='power.html', auto_open=False)
```

**–í—ã–≤–æ–¥:** –∫–∞–∫ –º—ã –≤–∏–¥–∏–º, –ø—Ä–∏ $\theta = \theta_0$ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —á–∞—Å—Ç–æ—Ç–∞ –æ—Ç–≤–µ—Ä–∂–µ–Ω–∏—è –≥–∏–ø–æ—Ç–µ–∑—ã –≤ —Ä–∞–π–æ–Ω–µ $0.05,$ (`count_of_samples` –±—ã–ª–æ –≤—ã–±—Ä–∞–Ω–æ 20, —á—Ç–æ–±—ã —Ç–∞–±–ª–∏—Ü–∞ –±—ã–ª–∞ –Ω–∞–≥–ª—è–¥–Ω–µ–µ) –∑–Ω–∞—á–∏—Ç, –∫—Ä–∏—Ç–µ—Ä–∏–π —Ä–∞–±–æ—Ç–∞–µ—Ç. –ö–∞—Å–∞–µ–º–æ –º–æ—â–Ω–æ—Å—Ç–∏, –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –ø—Ä–∏ $p \rightarrow 0.5$ –º–æ—â–Ω–æ—Å—Ç—å —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ $\alpha$ –ø—Ä–∏ –ª—é–±–æ–º $n,$ –∞ –ø—Ä–∏ $p \rightarrow 1$ –º–æ—â–Ω–æ—Å—Ç—å —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –µ–¥–∏–Ω–∏—Ü–µ –ø—Ä–∏ –ª—é–±–æ–º $n.$ –û–¥–Ω–∞–∫–æ, –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $(0.5, 1)$ –º—ã –≤–∏–¥–∏–º, —á—Ç–æ —Å —Ä–æ—Å—Ç–æ–º $n$ –º–æ—â–Ω–æ—Å—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏—è –≤—Å–µ –±—ã—Å—Ç—Ä–µ–µ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ 1.

## –ó–∞–¥–∞—á–∞ 2.

–ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ —á–∞—Å—Ç–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—é—Ç –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∫—Ä–∏—Ç–µ—Ä–∏–π –í–∞–ª—å–¥–∞. –¢–∞–∫–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏–∏ –∫–æ–Ω—Ç—Ä–æ–ª–∏—Ä—É—é—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏ I —Ä–æ–¥–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ $\alpha$ —Ç–æ–ª—å–∫–æ –≤ –ø—Ä–µ–¥–µ–ª–µ, –æ–¥–Ω–∞–∫–æ –¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö –≤—ã–±–æ—Ä–æ–∫ –æ–Ω–∞ –º–æ–∂–µ—Ç —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –æ—Ç $\alpha$, –ø—Ä–∏—á–µ–º –Ω–µ —Ä–µ–¥–∫–æ –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É. –ü–æ—ç—Ç–æ–º—É –¥–ª—è —Ç–∞–∫–∏—Ö –∫—Ä–∏—Ç–µ—Ä–∏–µ–≤ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç—å –≤ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ—à–∏–±–∫–∏ I —Ä–æ–¥–∞ –∏–ª–∏ *—Ä–µ–∞–ª—å–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏* –ø—Ä–∏ –∫–æ–Ω–µ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ.

–ó–∞—á–∞—Å—Ç—É—é –ø–æ—Å—á–∏—Ç–∞—Ç—å —Ç–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∞–ª—å–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –¥–æ–≤–æ–ª—å–Ω–æ —Å–ª–æ–∂–Ω–æ, –ø–æ—ç—Ç–æ–º—É –≤–º–µ—Å—Ç–æ –Ω–µ–≥–æ –≤—ã—á–∏—Å–ª—è—é—Ç –æ—Ü–µ–Ω–∫—É —Å –ø–æ–º–æ—â—å—é —Å–µ–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –º–µ—Ç–æ–¥—É –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ.

–ü—É—Å—Ç—å –≥–∏–ø–æ—Ç–µ–∑–∞ $\mathsf{H}_0$ –ø—Ä–æ—Å—Ç–∞—è, —Ç–æ –µ—Å—Ç—å $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. –î–ª—è –µ–µ –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø–æ –∫–æ–Ω–µ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ $X_1, ..., X_n$ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫—Ä–∏—Ç–µ—Ä–∏–π $S$. –û–ø–∏—à–∏—Ç–µ —Å—Ö–µ–º—É –æ—Ü–µ–Ω–∫–∏ —Ä–µ–∞–ª—å–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è $S$ –º–µ—Ç–æ–¥–æ–º –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ.

–ù–∞–ø–∏—à–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –æ—à–∏–±–∫–∏ –ø–µ—Ä–≤–æ–≥–æ —Ä–æ–¥–∞: $p(I_S) = P_{0}(x \in S).$ –≠—Ç—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –º–æ–∂–Ω–æ –ø—Ä–∏–±–ª–∏–∑–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫ –Ω—É–∂–Ω–æ–≥–æ –Ω–∞–º —Ä–∞–∑–º–µ—Ä–∞ –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $P_0$ –∏ —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É, —Å –∫–æ—Ç–æ—Ä–æ–π $x \in S.$ –î–∞–ª–µ–µ —É—Å—Ä–µ–¥–Ω—è–µ–º: —á–∞—Å—Ç–æ—Ç–∞ –≤ –ø—Ä–µ–¥–µ–ª–µ —Å–æ–π–¥–µ—Ç—Å—è –∫ –∏—Å–∫–æ–º–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $P_{0}(x \in S).$

–ï—Å–ª–∏ –≥–∏–ø–æ—Ç–µ–∑–∞ $\mathsf{H}_0$ —Å–ª–æ–∂–Ω–∞—è, —Ç–æ –µ—Å—Ç—å $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, —Ç–æ –º–æ–∂–Ω–æ –ø–æ–≤—Ç–æ—Ä–∏—Ç—å —Ç–∞–∫—É—é –ø—Ä–æ—Ü–µ–¥—É—Ä—É –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö $\mathsf{P} \in \mathscr{P}_0$.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—É $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. –û–ø–∏—à–∏—Ç–µ —Å—Ö–µ–º—É –æ—Ü–µ–Ω–∫–∏ –º–æ—â–Ω–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è $S$ –º–µ—Ç–æ–¥–æ–º –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ.

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ —Ç–æ–º—É, —á—Ç–æ –Ω–∞–ø–∏—Å–∞–Ω–æ –≤—ã—à–µ: –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫ –Ω—É–∂–Ω–æ–≥–æ –Ω–∞–º —Ä–∞–∑–º–µ—Ä–∞ –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $P_1$ –∏ —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É, —Å –∫–æ—Ç–æ—Ä–æ–π $x \in S.$ –î–∞–ª–µ–µ —É—Å—Ä–µ–¥–Ω—è–µ–º: —á–∞—Å—Ç–æ—Ç–∞ –≤ –ø—Ä–µ–¥–µ–ª–µ —Å–æ–π–¥–µ—Ç—Å—è –∫ –∏—Å–∫–æ–º–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $P_{1}(x \in S) = \beta_S(P_1).$

–ü—É—Å—Ç—å $X_1, ..., X_n$ &mdash; –≤—ã–±–æ—Ä–∫–∞ –∏–∑ –ø—É–∞—Å—Å–æ–Ω–æ–≤—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $\theta$. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—ã $\mathsf{H}\colon \theta = 1\ \ vs.\ \ \mathsf{H}\colon \theta \not= 1$ –∏ –∫—Ä–∏—Ç–µ—Ä–∏–π –í–∞–ª—å–¥–∞ –¥–ª—è –∏—Ö –ø—Ä–æ–≤–µ—Ä–∫–∏.

**1.** –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Ä–µ–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –¥–ª—è —Ä–∞–∑–º–µ—Ä–æ–≤ –≤—ã–±–æ—Ä–∫–∏ $1 \leqslant n \leqslant 50$ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å.

–î–ª—è –ø—É–∞—Å—Å–æ–Ω–æ–≤—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: $\widehat{\theta} = \overline{X}$ &mdash; –∞.–Ω.–æ. $\theta$ —Å –∞.–¥. $\sigma^2(\theta) = \theta.$ –¢–æ–≥–¥–∞ $\overline{X}$ &mdash; —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ $\theta$. –¢–æ–≥–¥–∞ $\widehat{\sigma}(\theta) = \sqrt{\overline{X}}.$ –ü—Ä–∏ $\theta = \theta_0$ –∫—Ä–∏—Ç–µ—Ä–∏–π –í–∞–ª—å–¥–∞ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: $W(X) = \sqrt{n}\frac{\widehat{\theta} - \theta_0}{\widehat{\sigma}}, \ S = \{|W(x)| \geq z_{1 - \alpha/2}\}$ —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –≤—ã—à–µ $\widehat{\theta}$ –∏ $\widehat{\sigma}.$ <br>
–£ –Ω–∞—Å –º–æ–∂–µ—Ç –≤–æ–∑–Ω–∏–∫–Ω—É—Ç—å –ø—Ä–æ–±–ª–µ–º–∞ —Å —Ç–µ–º, —á—Ç–æ $\widehat{\sigma} = 0.$ –≠—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç, –∫–æ–≥–¥–∞ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤—ã–±–æ—Ä–∫–∏ –Ω—É–ª–µ–≤—ã–µ. –ù–æ $0 \neq 1,$ –∞ $|W(x)|$ —É–ª–µ—Ç–∞–µ—Ç –≤ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç—å, –ø–æ—ç—Ç–æ–º—É —Ç—É—Ç –≤—Å–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ, –∏ –º—ã –¥–æ–ª–∂–Ω—ã –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å.

```python
Wald = lambda sigma, theta_hat, theta_0, n: np.sqrt(n) * (theta_hat - theta_0) / sigma
sigma_hat = lambda X: np.sqrt(np.average(X, axis=0))
count_frequency = lambda Y: np.average(Y, axis=0)
```

```python
theta_0 = 1
n_grid = np.linspace(1, 50, 50)
count_of_samples = 1000
alpha = 0.05
z_1_a2 = sps.norm.ppf(1 - alpha / 2)

frequency_arr = np.array([])
for n in n_grid:
    X = sps.poisson(mu=theta_0).rvs(size=(int(n), count_of_samples))
    W = Wald(sigma_hat(X), np.average(X, axis=0), theta_0, n)
    Y = [np.abs(W) > z_1_a2]
    frequency = count_frequency(Y)
    frequency = count_frequency(frequency)
    frequency_arr = np.append(frequency_arr, frequency)

fig = go.Figure()
fig.add_trace(go.Scatter(x=n_grid, y=frequency_arr, mode='lines'))
fig.update_layout(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ä–µ–∞–ª—å–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –æ—Ç ùëõ', 
                  xaxis_title="$n$", yaxis_title="$P_{I_S}(n)$")
fig.show()
pof.plot(fig, filename='significance_level.html', auto_open=False)
```

**–í—ã–≤–æ–¥:** –∫–∞–∫ –º—ã –≤–∏–¥–∏–º, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–π —Ä–µ–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏—Å–º–æ—Å—Ç–∏ –ø—Ä–∏ –º–∞–ª—ã—Ö —Ä–∞–∑–º–µ—Ä–∞—Ö –≤—ã–±–æ—Ä–∫–∏ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å –∂–µ–ª–∞–µ–º—ã–º $\alpha = 0.05.$ –ë–æ–ª–µ–µ —Ç–æ–≥–æ, –æ–Ω –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ –∂–µ–ª–∞–µ–º–æ–≥–æ, –ø–æ—ç—Ç–æ–º—É, –∫–∞–∫ –∏ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–ª–æ—Å—å, –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏–∏ —Ä–∞–±–æ—Ç–∞—é—Ç —Ç–æ–ª—å–∫–æ –¥–ª—è –±–æ–ª—å—à—ã—Ö —Ä–∞–∑–º–µ—Ä–æ–≤ –≤—ã–±–æ—Ä–∫–∏. –¢–∞–∫–∂–µ –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, –ø—Ä–∏ $n$ –ø–æ—Ä—è–¥–∫–∞ 50 —Ä–µ–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Å–æ—à–µ–ª—Å—è –∫ –Ω—É–∂–Ω–æ–º—É —É—Ä–æ–≤–Ω—é –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –Ω–æ –≤—Å–µ –µ—â–µ –∫–æ–ª–µ–±–ª–µ—Ç—Å—è –∑–∞—á–∞—Å—Ç—É—é –Ω–µ–º–Ω–æ–≥–æ –≤—ã—à–µ –µ–≥–æ.

**2.** –î–ª—è –∑–Ω–∞—á–µ–Ω–∏–π $n \in \{5, 15, 50\}$ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –º–æ—â–Ω–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –¥–ª—è –∑–Ω–∞—á–µ–Ω–∏–π $\theta \in (0, 5)$ –ø–æ —Å–µ—Ç–∫–µ —Å —à–∞–≥–æ–º 0.01 –ø–æ –º–µ—Ç–æ–¥—É –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ. –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Ç–∞–∫–∂–µ –º–æ—â–Ω–æ—Å—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏—è –ø–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–æ–π —Ñ–æ—Ä–º—É–ª–µ, –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –Ω–∞ –ª–µ–∫—Ü–∏–∏, –∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ —Å –æ—Ü–µ–Ω–∫–æ–π –º–æ—â–Ω–æ—Å—Ç–∏ –º–µ—Ç–æ–¥–æ–º –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ. –£—á—Ç–∏—Ç–µ, —á—Ç–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Ñ–æ—Ä–º—É–ª–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç $n$, –ø–æ—ç—Ç–æ–º—É –¥–ª—è –Ω–µ–µ —Ç–æ–∂–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ç—Ä–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.

–§–æ—Ä–º–∞–ª—å–Ω–æ –º–æ—â–Ω–æ—Å—Ç—å –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤ $\theta_0,$ –ø–æ—ç—Ç–æ–º—É –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤—ã–∫–æ–ª–µ–º —ç—Ç—É —Ç–æ—á–∫—É. <br>
–ü—Ä–∏ —Ä–∞—Å—á–µ—Ç–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –º–æ—â–Ω–æ—Å—Ç–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π, –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –Ω–∞ –ª–µ–∫—Ü–∏–∏: $\beta_S(\theta) = 1 - \Phi(z_{1 - \alpha/2} - W(\theta)) + \Phi(-z_{1 - \alpha/2} - W(\theta)).$

```python
theta_0 = 1
n_arr = np.array([5, 15, 50])
theta_grid = np.linspace(0.01, 4.99, 499)

frequency_arr = np.array([np.zeros(len(theta_grid))])
for n in n_arr:
    frequency_theta = np.array([])
    for theta in theta_grid:
        X = sps.poisson(mu=theta).rvs(size=(int(n), count_of_samples))
        W = Wald(sigma_hat(X), np.average(X, axis=0), theta_0, n)
        Y = [np.abs(W) > z_1_a2]
        frequency = count_frequency(Y)
        frequency = count_frequency(frequency)
        frequency_theta = np.append(frequency_theta, frequency)
    frequency_arr = np.append(frequency_arr, [frequency_theta], axis=0)
        
frequency_arr = frequency_arr[1:]
frequency_arr
```

```python
def count_theor_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.sf(z_1_a2 - W_theta) + sps.norm.cdf(-z_1_a2 - W_theta)

for i, n in enumerate(n_arr):
    frequency_theor = count_theor_beta(sigma=np.sqrt(theta_grid), theta=theta_grid, theta_0=theta_0, n=n)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=theta_grid, y=frequency_arr[i], mode='lines', 
                             name='–§—É–Ω–∫—Ü–∏—è –º–æ—â–Ω–æ—Å—Ç–∏ –ø–æ –º–µ—Ç–æ–¥—É –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ'))
    fig.add_trace(go.Scatter(x=[1], y=[frequency_arr[i, 99]], mode='markers', line_color='#FFFFFF', 
                             name='–í—ã–∫–æ–ª–æ—Ç–∞—è —Ç–æ—á–∫–∞'))
    fig.add_trace(go.Scatter(x=theta_grid, y=frequency_theor, mode='lines', 
                             name='–§—É–Ω–∫—Ü–∏—è –º–æ—â–Ω–æ—Å—Ç–∏, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–∞—è —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏'))
    fig.add_trace(go.Scatter(x=[1], y=[frequency_theor[99]], mode='markers', line_color='#FFFFFF', 
                             name='–í—ã–∫–æ–ª–æ—Ç–∞—è —Ç–æ—á–∫–∞'))
    fig.update_layout(title=f'–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –º–æ—â–Ω–æ—Å—Ç–∏ –æ—Ç ùúÉ –ø—Ä–∏ ùëõ = {n}', 
                      xaxis_title="$n$", yaxis_title="$\\beta_S(\\theta)$")
    fig.show()
    pof.plot(fig, filename=f'frequency_{n}.html', auto_open=False)
```

**–í—ã–≤–æ–¥:** –∫–∞–∫ –º—ã –≤–∏–¥–∏–º, –º–µ—Ç–æ–¥ –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ö–æ—Ä–æ—à–æ –ø—Ä–∏–±–ª–∏–∑–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –º–æ—â–Ω–æ—Å—Ç–∏. –û—Å–æ–±–µ–Ω–Ω–æ —Ö–æ—Ä–æ—à–æ —ç—Ç–æ –≤–∏–¥–Ω–æ –ø—Ä–∏ –±–æ–ª—å—à–∏—Ö $n$ (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ —ç—Ç–æ –ø—Ä–∏ $n = 50$). –¢–∞–∫–∂–µ –Ω–∞—à–∏ –≤—ã–≤–æ–¥—ã —Å 1 –∑–∞–¥–∞—á–∏ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–ª–∏—Å—å –∏ –¥–ª—è –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è: –ø—Ä–∏ $\theta \rightarrow \theta_0$ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å –ø—Ä–∏ –≤—Å–µ—Ö $n$ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ $\alpha,$ –∞ –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –æ—Ç $\theta_0$ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –º–æ—â–Ω–æ—Å—Ç—å —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –µ–¥–∏–Ω–∏—Ü–µ, –ø—Ä–∏ —ç—Ç–æ–º —Å —Ä–æ—Å—Ç–æ–º $n$ –≤—Å–µ —Å—Ç—Ä–µ–º–ª–µ–Ω–∏–µ –∫ –µ–¥–∏–Ω–∏—Ü–µ –≤—Å–µ –±—ã—Å—Ç—Ä–µ–µ –∏ –±—ã—Å—Ç—Ä–µ–µ.

**3.** –ü—É—Å—Ç—å $n=50$. –°—Ä–∞–≤–Ω–∏—Ç–µ –º–æ—â–Ω–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏–µ–≤, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π, –ª–µ–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∏ –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞–º.

–§–æ—Ä–º–∞–ª—å–Ω–æ, –∫–æ–Ω–µ—á–Ω–æ, —Å—Ç–æ–∏—Ç –¥–ª—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏—Ö –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤ –æ—Å—Ç–∞–≤–∏—Ç—å –º–æ—â–Ω–æ—Å—Ç—å —Ç–æ–ª—å–∫–æ —Ç–∞–º, –≥–¥–µ —Ñ—É–Ω–∫—Ü–∏—è —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –Ω–æ –Ω–∞–º –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, –∫–∞–∫ –∫—Ä–∏—Ç–µ—Ä–∏–∏ –í–∞–ª—å–¥–∞ –¥–ª—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏—Ö –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤ –≤–µ–¥—É—Ç —Å–µ–±—è –≤ —Ç–æ–π –æ–±–ª–∞—Å—Ç–∏, –≥–¥–µ –æ–Ω–∏ –Ω–µ –¥–æ–ª–∂–Ω—ã –æ—Ç–≤–µ—Ä–≥–∞—Ç—å.

```python
z_1_a = sps.norm.ppf(1 - alpha)
def count_theor_left_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.cdf(-z_1_a - W_theta)

def count_theor_right_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.sf(z_1_a - W_theta)
```

**–í–æ–ø—Ä–æ—Å –Ω–∞ –ø–æ–¥—É–º–∞—Ç—å.** –°–∫–æ–ª—å–∫–æ —Å–µ–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏–π –≤ –º–µ—Ç–æ–¥–µ –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å, —á—Ç–æ–±—ã –æ—Ü–µ–Ω–∏—Ç—å —Ä–µ–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ 0.001? –ò–Ω–∞—á–µ –≥–æ–≤–æ—Ä—è, –∑–Ω–∞—á–µ–Ω–∏—è (0.049, 0.051) –º–æ–∂–Ω–æ –ø—Ä–∏–Ω—è—Ç—å –∑–∞ 0.05.

–ó–∞–º–µ—á—É, —á—Ç–æ –Ω–∞ –ø–æ–¥—É–º–∞—Ç—å, –∞ –Ω–µ –Ω–∞ –ø–æ–ø–∏—Å–∞—Ç—å –∫–æ–¥, –ø–æ—ç—Ç–æ–º—É –µ—Å–ª–∏ –∏ –∑–∞ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ–∫ –∞–Ω–∞–ª–∏–∑–∞ —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ –∑–¥–µ—Å—å —Å–Ω–∏–º—É—Ç –±–∞–ª–ª—ã, —Ç–æ —ç—Ç–æ –±—É–¥–µ—Ç —É–∂–µ —Å–æ–≤—Å–µ–º –Ω–µ–∫—Ä—É—Ç–æ :) <br>
–ü—É—Å—Ç—å —É –Ω–∞—Å $m$ –≤—ã–±–æ—Ä–æ–∫ –ø–æ $n$ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–π. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤—ã–±–æ—Ä–∫—É –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –∫–∞–∫ –æ–¥–Ω—É –±–æ–ª—å—à—É—é –≤—ã–±–æ—Ä–∫—É. –ü–æ –¶–ü–¢ $W(X) = \sqrt{nm}\frac{\widehat{\theta} - \theta_0}{\widehat{\sigma}} \stackrel{_{d_{\theta_0}}}{\rightarrow}N(0, 1).$ –¢–æ–≥–¥–∞ –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –º—ã —Å—Ö–æ–¥–∏–º—Å—è –∫ –Ω—É–∂–Ω–æ–º—É $\alpha$ —Å–æ —Å–∫–æ—Ä–æ—Å—Ç—å—é –ø—Ä–∏–º–µ—Ä–Ω–æ $\sqrt{nm}.$ –†–∞—Å—Å—á–∏—Ç–∞–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—É—é –æ—à–∏–±–∫—É, –∫–æ—Ç–æ—Ä—É—é –º—ã –º–æ–∂–µ–º –¥–æ–ø—É—Å—Ç–∏—Ç—å: $\varepsilon = \frac{\Delta \alpha}{\alpha} = 0.02.$ –†–∞—Å—Å—á–∏—Ç–∞–µ–º $nm$ –∏–∑ —É—Å–ª–æ–≤–∏—è: $\sqrt{nm} = \frac{1}{\varepsilon},$ –æ—Ç–∫—É–¥–∞ $nm = 2500.$ –í–æ–∑—å–º–µ–º $n = 50,$ —Ç–æ–≥–¥–∞ –∏—Å–∫–æ–º–æ–µ $m = 50.$ –ü–æ—Å–º–æ—Ç—Ä–µ–≤ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫, –≤–∏–¥–∏–º, —á—Ç–æ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –º—ã –Ω–µ —Å–æ—à–ª–∏—Å—å —Å –Ω—É–∂–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é –ø—Ä–∏ $n = 50$ –¥–∞–∂–µ –ø—Ä–∏ $m = 200,$ –æ–¥–Ω–∞–∫–æ –ø–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ —Ä–∞—Å—Å—É–∂–¥–µ–Ω–∏—è –±–µ–∑ –ø–æ–ø—ã—Ç–∫–∏ —á—Ç–æ-—Ç–æ —Å—Ç—Ä–æ–≥–æ –¥–æ–∫–∞–∑–∞—Ç—å.

–ü—É—Å—Ç—å $X_1, ..., X_n$ &mdash; –≤—ã–±–æ—Ä–∫–∞ –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\Gamma(\theta, \beta)$, –ø—Ä–∏—á–µ–º $\beta$ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ—Å–Ω–æ–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—É $\mathsf{H}\colon \theta = 2$.

**1.** –î–ª—è —Ä–∞–∑–º–µ—Ä–æ–≤ –≤—ã–±–æ—Ä–∫–∏ $n \in \{5, 15, 30, 50, 100\}$ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Ä–µ–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –í–∞–ª—å–¥–∞ –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏–π $\beta$. –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.

–ò–∑ –¥/–∑ –º—ã –≤—ã—è—Å–Ω–∏–ª–∏, —á—Ç–æ –∫—Ä–∏—Ç–µ—Ä–∏–π –í–∞–ª—å–¥–∞ –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã –ø—Ä–æ–≤–µ—Ä–∫–∏ –æ—Å–Ω–æ–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã $\theta = \theta_0$ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:
    $$W(X) = \frac{\sqrt{n}}{\sqrt{\frac{2\overline{X}^2}{S^4} + \frac{3}{S^2}}}\left(\frac{\overline{X}}{S^2} - \theta_0 \right), \ S_1 = \{|W(x)| \geq z_{1 - \alpha/2}\}.$$

```python
sigma_est_gamma = lambda X_avg, S: np.sqrt(2 * (X_avg ** 2) / (S ** 2) + 3 / S)
theta_est_gamma = lambda X_avg, S: X_avg / S
```

**2.** –°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ –º–æ—â–Ω–æ—Å—Ç–∏ –†–ù–ú–ö –∏ –∫—Ä–∏—Ç–µ—Ä–∏–π –í–∞–ª—å–¥–∞ –¥–ª—è –ª–µ–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∏ –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤, –∞ —Ç–∞–∫–∂–µ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã –¥–ª—è –∫—Ä–∏—Ç–µ—Ä–∏—è –í–∞–ª—å–¥–∞.

–í —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º –¥/–∑ –º—ã —Ä–∞—Å—Å—á–∏—Ç–∞–ª–∏ –º–æ—â–Ω–æ—Å—Ç—å –†–ù–ú–ö $(T(X) = \sum\limits_{i = 1}^{n}X_i)$: –¥–ª—è –ª–µ–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã: 
$$1 - F_{T(X), \theta_0}(c_{\alpha}) = \alpha,$$ –æ—Ç–∫—É–¥–∞ $c_{\alpha} = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta_0}).$ `ppf` $(1 - \alpha).$ <br>
–ú–æ—â–Ω–æ—Å—Ç—å: $\beta_S(\theta) = P_{\theta}(T(X) \geq c_{\alpha}) = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta}).$ `sf` $(c_{\alpha}).$ <br>
–î–ª—è –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã:
$$F_{T(X), \theta_0}(c_{\alpha}) = \alpha,$$ –æ—Ç–∫—É–¥–∞ $c_{\alpha} = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta_0}).$ `ppf` $(\alpha).$ <br>
–ú–æ—â–Ω–æ—Å—Ç—å: $\beta_S(\theta) = P_{\theta}(T(X) \leq c_{\alpha}) = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta}).$ `cdf` $(c_{\alpha}).$ <br>
–û–¥–Ω–∞–∫–æ —á—Ç–æ –¥–µ–ª–∞—Ç—å, –µ—Å–ª–∏ $c_{\alpha}$ –∏–∑–≤–µ—Å—Ç–Ω–æ, –æ–¥–Ω–∞–∫–æ –º—ã –Ω–µ –º–æ–∂–µ–º —Ç–∞–∫ –ø—Ä–æ—Å—Ç–æ –≤–∑—è—Ç—å `sf` –∏–ª–∏ `cdf`? –¢—É—Ç –Ω–∞–º —Å–Ω–æ–≤–∞ –ø—Ä–∏—Ö–æ–¥–∏—Ç –Ω–∞ –ø–æ–º–æ—â—å –º–µ—Ç–æ–¥ –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ. –ü–æ–ø—Ä–æ–±—É–µ–º —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –º–æ—â–Ω–æ—Å—Ç—å –†–ú–ù–ö —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ.

```python
def left_RMNK_c_a(n, beta, theta_0):
    return sps.gamma(a=n * beta, scale=1 / theta_0).ppf(1 - alpha) 

def right_RMNK_c_a(n, beta, theta_0):
    return sps.gamma(a=n * beta, scale=1 / theta_0).ppf(alpha)
```

```python
count_of_samples = 5000
theta_0 = 2
n_arr = np.array([5, 15, 30, 50, 100])
theta_grid = np.linspace(0.01, 10, 1000)
beta_0 = 2

frequency_arr = np.array([np.zeros(len(theta_grid))])
frequency_left_arr = np.array([np.zeros(len(theta_grid))])
frequency_right_arr = np.array([np.zeros(len(theta_grid))])
left_RMNK_arr = np.array([np.zeros(len(theta_grid))])
right_RMNK_arr = np.array([np.zeros(len(theta_grid))])
for n in n_arr:
    frequency_theta = np.array([])
    left_frequency_theta = np.array([])
    right_frequency_theta = np.array([])
    left_RMNK_theta = np.array([])
    right_RMNK_theta = np.array([])
    for theta in theta_grid:
        X = sps.gamma(a=beta_0, scale=1/theta).rvs(size=(int(n), count_of_samples))
        
        W = Wald(sigma_est_gamma(np.average(X, axis=0), np.var(X, axis=0)), 
                 theta_est_gamma(np.average(X, axis=0), np.var(X, axis=0)), theta_0, n)
        left_c_a = left_RMNK_c_a(n, beta_0, theta_0)
        right_c_a = right_RMNK_c_a(n, beta_0, theta_0)
        
        Y = [np.abs(W) > z_1_a2]
        Y_left = [W < -z_1_a]
        Y_right = [W > z_1_a]
        Y_RMNK_left = [np.sum(X, axis=0) > left_c_a]
        Y_RMNK_right = [np.sum(X, axis=0) < right_c_a]
        
        frequency = count_frequency(Y)
        frequency = count_frequency(frequency)
        left_frequency = count_frequency(Y_left)
        left_frequency = count_frequency(left_frequency)
        right_frequency = count_frequency(Y_right)
        right_frequency = count_frequency(right_frequency)
        left_RMNK_frequency = count_frequency(Y_RMNK_left)
        left_RMNK_frequency = count_frequency(left_RMNK_frequency)
        right_RMNK_frequency = count_frequency(Y_RMNK_right)
        right_RMNK_frequency = count_frequency(right_RMNK_frequency)
        
        frequency_theta = np.append(frequency_theta, frequency)
        left_frequency_theta = np.append(left_frequency_theta, left_frequency)
        right_frequency_theta = np.append(right_frequency_theta, right_frequency)
        left_RMNK_theta = np.append(left_RMNK_theta, left_RMNK_frequency)
        right_RMNK_theta = np.append(right_RMNK_theta, right_RMNK_frequency)
        
    frequency_arr = np.append(frequency_arr, [frequency_theta], axis=0)
    frequency_left_arr = np.append(frequency_left_arr, [left_frequency_theta], axis=0)
    frequency_right_arr = np.append(frequency_right_arr, [right_frequency_theta], axis=0)
    left_RMNK_arr = np.append(left_RMNK_arr, [left_RMNK_theta], axis=0)
    right_RMNK_arr = np.append(right_RMNK_arr, [right_RMNK_theta], axis=0)
    
frequency_arr = frequency_arr[1:]
frequency_left_arr = frequency_left_arr[1:]
frequency_right_arr = frequency_right_arr[1:]
left_RMNK_arr = left_RMNK_arr[1:]
right_RMNK_arr = right_RMNK_arr[1:]
right_RMNK_arr
```
