## Задача 1.

Существует примета, что если перед вами дорогу перебегает черный кот, то скоро случится неудача. 
Вы же уже достаточно хорошо знаете статистику и хотите проверить данную примету. 
Сформулируем задачу на математическом языке. 

Пусть $X_1, ..., X_n \sim Bern(p)$ &mdash; проведенные наблюдения, где $X_i = 1$, если в $i$-м испытании случилась неудача после того, как черный кот перебежал дорогу, а $p$ &mdash; неизвестная вероятность такого события. 
Вы хотите проверить гипотезу $\mathsf{H}_0\colon p=1/2$ против альтернативы $\mathsf{H}_1\colon p>1/2$. Нулевая гипотеза означает отсутствие связи между черным котом и неудачей, первая &mdash; что неудача происходит чаще если черный кот перебегает дорогу.

На семинаре получено, что $S = \left\{T(x) \geqslant c_\alpha\right\}$, где $T(X) = \sum\limits_{i=1}^n X_i$, является равномерно наиболее мощным критерием для проверки этих гипотез. 
Чему при этом равно $c_\alpha$?

 На семинарах мы выяснили, что $$c_{\alpha} = \min_{k \in \mathbb{N}}\sum\limits_{i = 1}^{k}p_{n, 1/2}(k) \geq \alpha = $$ $ = $ `sps.binom(n, 1/2).ppf`($1 - \alpha$).

Проверьте, что критерий работает.
Возьмите несколько значений $n$ и реализаций статистики $T(X)$.
В каждом случае найдите значение $c_\alpha$, а также результат процедуры проверки гипотез 0/1 &mdash; отвергается ли $\mathsf{H}_0$ (выводите 1, если отвергается).
Оформите это в виде таблицы, можно через `pandas.DataFrame`.

*Замечание:* Пользуйтесь функциями из `scipy.stats`.

```python
alpha = 0.05
p = 0.5
count_of_samples = 20
n_arr = np.array([5, 20, 100, 1000, 10000])


c_a = np.array([])
frequency_reject = np.array([])
reject_arr = np.array([np.zeros(count_of_samples)])
for n in n_arr:
    c_a = np.append(c_a, sps.binom(n, p).ppf(1 - alpha))
    t = sps.binom(n, p).rvs(size=count_of_samples)
    reject = [t > c_a[-1]]
    reject += np.zeros(len(reject))
    reject_arr = np.append(reject_arr, reject, axis=0)
    frequency_reject = np.append(frequency_reject, np.sum(reject))

frequency_reject /= count_of_samples    
reject_arr = reject_arr[1:]
reject_arr
```

Формула для мощности: $\beta_S(\theta) = P_{\theta}(T(x) \geq c_\alpha)$

```python
power = lambda p, n, c_a: sps.binom(n, p).sf(c_a)

p_grid = np.linspace(0.501, 0.999, 499)
fig = go.Figure()
for i, n in enumerate(n_arr):
    power_grid = power(p_grid, n, c_a[i])
    fig.add_trace(go.Scatter(x=p_grid, y=power_grid, mode='lines', name=f'𝑛 = {n}'))
fig.update_layout(title='Функция мощности критерия для разных 𝑛', 
                  xaxis_title="$p$", yaxis_title="$\\beta_{S}(p)$")
fig.show()
pof.plot(fig, filename='power.html', auto_open=False)
```

## Задача 2.

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем не редко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью семплирования по методу Монте-Карло.

Пусть гипотеза $\mathsf{H}_0$ простая, то есть $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. Для ее проверки по конечной выборке $X_1, ..., X_n$ используется критерий $S$. Опишите схему оценки реального уровня значимости критерия $S$ методом Монте-Карло.

Напишем выражение для вычисления ошибки первого рода: $p(I_S) = P_{0}(x \in S).$ Эту вероятность можно приблизить следующим образом: генерируем большое количество выборок нужного нам размера из распределения $P_0$ и считаем частоту, с которой $x \in S.$ Далее усредняем: частота в пределе сойдется к искомой вероятности $P_{0}(x \in S).$

Если гипотеза $\mathsf{H}_0$ сложная, то есть $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, то можно повторить такую процедуру несколько раз для разных $\mathsf{P} \in \mathscr{P}_0$.

Рассмотрим альтернативную гипотезу $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. Опишите схему оценки мощности критерия $S$ методом Монте-Карло.

Аналогично тому, что написано выше: генерируем большое количество выборок нужного нам размера из распределения $P_1$ и считаем частоту, с которой $x \in S.$ Далее усредняем: частота в пределе сойдется к искомой вероятности $P_{1}(x \in S) = \beta_S(P_1).$

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}\colon \theta = 1\ \ vs.\ \ \mathsf{H}\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** Посчитайте реальный уровень значимости критерия для размеров выборки $1 \leqslant n \leqslant 50$ и визуализируйте полученную зависимость.

Для пуассоновского распределения: $\widehat{\theta} = \overline{X}$ &mdash; а.н.о. $\theta$ с а.д. $\sigma^2(\theta) = \theta.$ Тогда $\overline{X}$ &mdash; состоятельная оценка $\theta$. Тогда $\widehat{\sigma}(\theta) = \sqrt{\overline{X}}.$ При $\theta = \theta_0$ критерий Вальда выглядит следующим образом: $W(X) = \sqrt{n}\frac{\widehat{\theta} - \theta_0}{\widehat{\sigma}}, \ S = \{|W(x)| \geq z_{1 - \alpha/2}\}$ с указанными выше $\widehat{\theta}$ и $\widehat{\sigma}.$ <br>
У нас может возникнуть проблема с тем, что $\widehat{\sigma} = 0.$ Это происходит, когда все элементы выборки нулевые. Но $0 \neq 1,$ а $|W(x)|$ улетает в бесконечность, поэтому тут все работает правильно, и мы должны отвергнуть.

```python
Wald = lambda sigma, theta_hat, theta_0, n: np.sqrt(n) * (theta_hat - theta_0) / sigma
sigma_hat = lambda X: np.sqrt(np.average(X, axis=0))
count_frequency = lambda Y: np.average(Y, axis=0)
```

```python
theta_0 = 1
n_grid = np.linspace(1, 50, 50)
count_of_samples = 1000
alpha = 0.05
z_1_a2 = sps.norm.ppf(1 - alpha / 2)

frequency_arr = np.array([])
for n in n_grid:
    X = sps.poisson(mu=theta_0).rvs(size=(int(n), count_of_samples))
    W = Wald(sigma_hat(X), np.average(X, axis=0), theta_0, n)
    Y = [np.abs(W) > z_1_a2]
    frequency = count_frequency(Y)
    frequency = count_frequency(frequency)
    frequency_arr = np.append(frequency_arr, frequency)

fig = go.Figure()
fig.add_trace(go.Scatter(x=n_grid, y=frequency_arr, mode='lines'))
fig.update_layout(title='Зависимость реального уровня значимости от 𝑛', 
                  xaxis_title="$n$", yaxis_title="$P_{I_S}(n)$")
fig.show()
pof.plot(fig, filename='significance_level.html', auto_open=False)
```

**Вывод:** как мы видим, действительно, рассчитанный реальный уровень значисмости при малых размерах выборки не совпадает с желаемым $\alpha = 0.05.$ Более того, он значительно больше желаемого, поэтому, как и предполагалось, асимптотические критерии работают только для большых размеров выборки. Также можно заметить, при $n$ порядка 50 реальный уровень значимости сошелся к нужному уровню значимости, но все еще колеблется зачастую немного выше его.

**2.** Для значений $n \in \{5, 15, 50\}$ посчитайте функцию мощности критерия для значений $\theta \in (0, 5)$ по сетке с шагом 0.01 по методу Монте-Карло. Визуализируйте полученные зависимости. Посчитайте также мощность критерия по асимптотической формуле, полученной на лекции, и сравните с оценкой мощности методом Монте-Карло. Учтите, что асимптотическая формула содержит $n$, поэтому для нее тоже должно быть три зависимости.

Формально мощность не определена в $\theta_0,$ поэтому на графике выколем эту точку. <br>
При расчете теоретической мощности воспользуемся формулой, полученной на лекции: $\beta_S(\theta) = 1 - \Phi(z_{1 - \alpha/2} - W(\theta)) + \Phi(-z_{1 - \alpha/2} - W(\theta)).$

```python
theta_0 = 1
n_arr = np.array([5, 15, 50])
theta_grid = np.linspace(0.01, 4.99, 499)

frequency_arr = np.array([np.zeros(len(theta_grid))])
for n in n_arr:
    frequency_theta = np.array([])
    for theta in theta_grid:
        X = sps.poisson(mu=theta).rvs(size=(int(n), count_of_samples))
        W = Wald(sigma_hat(X), np.average(X, axis=0), theta_0, n)
        Y = [np.abs(W) > z_1_a2]
        frequency = count_frequency(Y)
        frequency = count_frequency(frequency)
        frequency_theta = np.append(frequency_theta, frequency)
    frequency_arr = np.append(frequency_arr, [frequency_theta], axis=0)
        
frequency_arr = frequency_arr[1:]
frequency_arr
```

```python
def count_theor_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.sf(z_1_a2 - W_theta) + sps.norm.cdf(-z_1_a2 - W_theta)

for i, n in enumerate(n_arr):
    frequency_theor = count_theor_beta(sigma=np.sqrt(theta_grid), theta=theta_grid, theta_0=theta_0, n=n)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=theta_grid, y=frequency_arr[i], mode='lines', 
                             name='Функция мощности по методу Монте-Карло'))
    fig.add_trace(go.Scatter(x=[1], y=[frequency_arr[i, 99]], mode='markers', line_color='#FFFFFF', 
                             name='Выколотая точка'))
    fig.add_trace(go.Scatter(x=theta_grid, y=frequency_theor, mode='lines', 
                             name='Функция мощности, рассчитанная теоретически'))
    fig.add_trace(go.Scatter(x=[1], y=[frequency_theor[99]], mode='markers', line_color='#FFFFFF', 
                             name='Выколотая точка'))
    fig.update_layout(title=f'Зависимость функции мощности от 𝜃 при 𝑛 = {n}', 
                      xaxis_title="$n$", yaxis_title="$\\beta_S(\\theta)$")
    fig.show()
    pof.plot(fig, filename=f'frequency_{n}.html', auto_open=False)
```

**Вывод:** как мы видим, метод Монте-Карло позволяет хорошо приблизить функцию мощности. Особенно хорошо это видно при больших $n$ (в нашем случае это при $n = 50$). Также наши выводы с 1 задачи подтвердились и для асимптотического критерия: при $\theta \rightarrow \theta_0$ теоретически рассчитанная мощность при всех $n$ стремится к $\alpha,$ а при удалении от $\theta_0$ теоретическая мощность стремится к единице, при этом с ростом $n$ все стремление к единице все быстрее и быстрее.

**3.** Пусть $n=50$. Сравните мощности критериев, соответствующих двусторонней, левосторонней и правосторонней альтернативам.

Формально, конечно, стоит для односторонних альтернатив оставить мощность только там, где функция существует, но нам интересно посмотреть, как критерии Вальда для односторонних альтернатив ведут себя в той области, где они не должны отвергать.

```python
z_1_a = sps.norm.ppf(1 - alpha)
def count_theor_left_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.cdf(-z_1_a - W_theta)

def count_theor_right_beta(sigma, theta, theta_0, n): 
    W_theta = Wald(sigma, theta, theta_0, n)
    return sps.norm.sf(z_1_a - W_theta)
```

```python
n = 50

frequency_theor = count_theor_beta(sigma=np.sqrt(theta_grid), theta=theta_grid, theta_0=theta_0, n=n)
frequency_theor_left = count_theor_left_beta(sigma=np.sqrt(theta_grid), theta=theta_grid, theta_0=theta_0, n=n)
frequency_theor_right = count_theor_right_beta(sigma=np.sqrt(theta_grid), theta=theta_grid, theta_0=theta_0, n=n)

fig = go.Figure()
fig.add_trace(go.Scatter(x=theta_grid, y=frequency_theor, mode='lines', 
                             name='Функция мощности двусторонней альтернативы'))
fig.add_trace(go.Scatter(x=[1], y=[frequency_theor[100]], mode='markers', line_color='#FFFFFF', 
                             name='Выколотая точка'))
fig.add_trace(go.Scatter(x=theta_grid, y=frequency_theor_left, mode='lines', 
                             name='Функция мощности левосторонней альтернативы'))
fig.add_trace(go.Scatter(x=[1], y=[frequency_theor[100]], mode='markers', line_color='#FFFFFF', 
                             name='Выколотая точка'))
fig.add_trace(go.Scatter(x=theta_grid, y=frequency_theor_right, mode='lines', 
                             name='Функция мощности правосторонней альтернативы'))
fig.add_trace(go.Scatter(x=[1], y=[frequency_theor[100]], mode='markers', line_color='#FFFFFF', 
                             name='Выколотая точка'))
fig.update_layout(title=f'Зависимость функции мощности от 𝜃 при 𝑛 = {n}', 
                      xaxis_title="$n$", yaxis_title="$\\beta_S(\\theta)$")
fig.show()
pof.plot(fig, filename=f'frequency_left_right.html', auto_open=False)
```

**Вывод:** как видно из графиков, все функции мощности стремится к $\alpha,$ при $\theta \rightarrow \theta_0.$ Также заметим, что, как и предполагалось, правосторонняя альтернатива мощнее всех при $\theta > \theta_0$ и имеет "мощность", близкую к нулю при $\theta < \theta_0,$ в то время как левосторонняя альтернатива мощнее всех при $\theta < \theta_0$ и имеет "мощность", близкую к нулю при $\theta > \theta_0.$ Двусторонняя альтернатива же отвергает и при $\theta < \theta_0,$ и при $\theta > \theta_0$ с высокой вероятностью, но чуть меньшей, чем левосторонняя и правосторонняя альтернатива соответственно.

Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\Gamma(\theta, \beta)$, причем $\beta$ неизвестно. Рассмотрим основную гипотезу $\mathsf{H}\colon \theta = 2$.

    $$W(X) = \frac{\sqrt{n}}{\sqrt{\frac{2\overline{X}^2}{S^4} + \frac{3}{S^2}}}\left(\frac{\overline{X}}{S^2} - \theta_0 \right), \ S_1 = \{|W(x)| \geq z_{1 - \alpha/2}\}.$$

**2.** Сравните по мощности РНМК и критерий Вальда для левосторонней и правосторонней альтернатив, а также двусторонней альтернативы для критерия Вальда.

В теоретическом д/з мы рассчитали мощность РНМК $(T(X) = \sum\limits_{i = 1}^{n}X_i)$: для левосторонней альтернативы: 
$$1 - F_{T(X), \theta_0}(c_{\alpha}) = \alpha,$$ откуда $c_{\alpha} = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta_0}).$ `ppf` $(1 - \alpha).$ <br>
Мощность: $\beta_S(\theta) = P_{\theta}(T(X) \geq c_{\alpha}) = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta}).$ `sf` $(c_{\alpha}).$ <br>
Для правосторонней альтернативы:
$$F_{T(X), \theta_0}(c_{\alpha}) = \alpha,$$ откуда $c_{\alpha} = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta_0}).$ `ppf` $(\alpha).$ <br>
Мощность: $\beta_S(\theta) = P_{\theta}(T(X) \leq c_{\alpha}) = $ `sps.gamma` $(a = n\beta, scale = \frac{1}{\theta}).$ `cdf` $(c_{\alpha}).$ <br>
Однако что делать, если $c_{\alpha}$ известно, однако мы не можем так просто взять `sf` или `cdf`? Тут нам снова приходит на помощь метод Монте-Карло. Попробуем рассчитать мощность РМНК с помощью метода Монте-Карло.

```python
def left_RMNK_c_a(n, beta, theta_0):
    return sps.gamma(a=n * beta, scale=1 / theta_0).ppf(1 - alpha) 

def right_RMNK_c_a(n, beta, theta_0):
    return sps.gamma(a=n * beta, scale=1 / theta_0).ppf(alpha)
```

```python
count_of_samples = 5000
theta_0 = 2
n_arr = np.array([5, 15, 30, 50, 100])
theta_grid = np.linspace(0.01, 10, 1000)
beta_0 = 2

frequency_arr = np.array([np.zeros(len(theta_grid))])
frequency_left_arr = np.array([np.zeros(len(theta_grid))])
frequency_right_arr = np.array([np.zeros(len(theta_grid))])
left_RMNK_arr = np.array([np.zeros(len(theta_grid))])
right_RMNK_arr = np.array([np.zeros(len(theta_grid))])
for n in n_arr:
    frequency_theta = np.array([])
    left_frequency_theta = np.array([])
    right_frequency_theta = np.array([])
    left_RMNK_theta = np.array([])
    right_RMNK_theta = np.array([])
    for theta in theta_grid:
        X = sps.gamma(a=beta_0, scale=1/theta).rvs(size=(int(n), count_of_samples))
        
        W = Wald(sigma_est_gamma(np.average(X, axis=0), np.var(X, axis=0)), 
                 theta_est_gamma(np.average(X, axis=0), np.var(X, axis=0)), theta_0, n)
        left_c_a = left_RMNK_c_a(n, beta_0, theta_0)
        right_c_a = right_RMNK_c_a(n, beta_0, theta_0)
        
        Y = [np.abs(W) > z_1_a2]
        Y_left = [W < -z_1_a]
        Y_right = [W > z_1_a]
        Y_RMNK_left = [np.sum(X, axis=0) > left_c_a]
        Y_RMNK_right = [np.sum(X, axis=0) < right_c_a]
        
        frequency = count_frequency(Y)
        frequency = count_frequency(frequency)
        left_frequency = count_frequency(Y_left)
        left_frequency = count_frequency(left_frequency)
        right_frequency = count_frequency(Y_right)
        right_frequency = count_frequency(right_frequency)
        left_RMNK_frequency = count_frequency(Y_RMNK_left)
        left_RMNK_frequency = count_frequency(left_RMNK_frequency)
        right_RMNK_frequency = count_frequency(Y_RMNK_right)
        right_RMNK_frequency = count_frequency(right_RMNK_frequency)
        
        frequency_theta = np.append(frequency_theta, frequency)
        left_frequency_theta = np.append(left_frequency_theta, left_frequency)
        right_frequency_theta = np.append(right_frequency_theta, right_frequency)
        left_RMNK_theta = np.append(left_RMNK_theta, left_RMNK_frequency)
        right_RMNK_theta = np.append(right_RMNK_theta, right_RMNK_frequency)
        
    frequency_arr = np.append(frequency_arr, [frequency_theta], axis=0)
    frequency_left_arr = np.append(frequency_left_arr, [left_frequency_theta], axis=0)
    frequency_right_arr = np.append(frequency_right_arr, [right_frequency_theta], axis=0)
    left_RMNK_arr = np.append(left_RMNK_arr, [left_RMNK_theta], axis=0)
    right_RMNK_arr = np.append(right_RMNK_arr, [right_RMNK_theta], axis=0)
    
frequency_arr = frequency_arr[1:]
frequency_left_arr = frequency_left_arr[1:]
frequency_right_arr = frequency_right_arr[1:]
left_RMNK_arr = left_RMNK_arr[1:]
right_RMNK_arr = right_RMNK_arr[1:]
right_RMNK_arr
```

Критерий Вальда двусторонней альтернативы, очевидно, при малых $n$ тоже несодержателен справа, однако слева показывает при всех $n$ довольно неплохие результаты. Во всех случаях ожидаемо самыми мощными оказались РМНК на участках, на которых они и должны отвергать. При этом заметно, что с ростом $n$ все критерии начинают работать лучше. При выборе между критериями Вальда односторонней или двусторонней альтернативы стоит выбирать тот, который соответствует рассматриваемой альтернативе: на участке $\theta > \theta_0$ лучше всего отвергает правосторонний, его и будем брать, на участке $\theta < \theta_0$ лучше всего отвергает левосторонний, его и будем брать, а если у нас альтернатива $\theta \neq \theta_0,$ то хорошо отвергает с обеих сторон двусторонний критерий Вальда.
